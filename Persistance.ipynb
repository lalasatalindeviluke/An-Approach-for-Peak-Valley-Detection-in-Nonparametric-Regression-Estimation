```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
peak_x = [0, 2, 4, 6, 8, 10, 12]
peak_y = [1.5, 1.2, 0.4, 0.43, 0.23, 0.16, 0.18]

valley_x = [1, 3, 5, 7, 9, 11, 13]
valley_y = [0.8, 0.2, 0.385, 0.215, 0.15, 0.15, 0]
```

```python
table = pd.DataFrame({
            "X": peak_x+valley_x,
            "Y": peak_y+valley_y
        })
table = table.sort_values("X")
table
```

```python
persistance_x = [1.5, 1.2, 0.4, 0.43] # [1.5, 1.2, 0.43, 0.18, 0.4, 0.23, 0.16]
persistance_y = [0, 0.8, 0.38, 0.2] # [0, 0.8, 0.2, 0.15, 0.38, 0.215, 0.15]

annotation_x = [0, 2, 4, 6]
annotation_y = [1.5, 1.2, 0.4, 0.43]
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(16, 9), dpi=300)

ax1.plot(table["X"], table["Y"])
ax1.scatter(peak_x, peak_y, label="Top", marker="x", c='r')
ax1.scatter(valley_x, valley_y, label="Bottom", marker="o", c='orange')
for i in range(4):
    ax1.text(annotation_x[i], annotation_y[i], f"{i+1}", size=20)
ax1.legend(fontsize=15)
ax1.grid()

ax2.plot([0, 1.6], [0, 1.6], c="k")
ax2.scatter(persistance_x, persistance_y, s=16, c="r")
for i in range(4):
    ax2.text(persistance_x[i], persistance_y[i], f"{i+1}", size=20)
ax2.grid()
ax2.set_xlim(0, 1.6)
ax2.set_ylim(0, 1.6)
ax2.set_xlabel("Birth Level", fontsize=25)
ax2.set_ylabel("Death Level", fontsize=25)
# ax2.scatter()

```

```python
def persistence(x: list, y: list) -> np.array:
    '''
        x - y = 0
    '''
    return np.abs(1*np.array(x)-1*np.array(y)) / np.sqrt(1**2 + (-1)**2)
```

```python
persistence(persistance_x, persistance_y)
```
