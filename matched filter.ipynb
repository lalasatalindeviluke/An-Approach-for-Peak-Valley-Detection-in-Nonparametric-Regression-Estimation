```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

$$ f(t) = \frac{1}{\sqrt{2\pi}}\exp{\frac{-t^2}{2}} $$

```python
t = Symbol('t')
t1 = Symbol('t1')
f = (1/sqrt(2*pi))*exp(-t**2/2)
# f = (1/sqrt(2*pi))*exp(-t**2/2) + (1/(sqrt(2*pi)*sqrt(0.5)))*exp(-(t-2.5)**2/(2*0.5))
```

```python
f1 = diff(f, t)
f1
```

```python
beta = f * diff(f, t, 3).evalf(subs={'t':t+t1})
beta
```

```python
beta_t = integrate(beta, (t, -oo, oo))
# beta_t
```

```python
plot(beta_t.evalf())
plt.show()
```

```python
alpha_numerator = f * diff(f, t, 2).evalf(subs={'t':t+t1})
alpha_denominator = (diff(f, t))**2
```

```python
alpha_t = integrate(alpha_numerator, (t, -oo, oo)) / integrate(alpha_denominator, (t, -oo, oo))
# alpha_t
```

```python
plot(alpha_t.evalf())
plt.show()
```

```python
f2 = diff(f, t, 2)
f2.evalf(subs={'t':t+t1})
```

```python
f3 = diff(f, t, 3)
f3.evalf(subs={'t':t+t1})
```

```python
(f3/4).evalf(subs={'t': t1/sqrt(2)})
```

```python
plot(f.evalf(subs={'t':t1}), adaptive=False, nb_of_points=1000)
plt.show()
```

```python
plot((f3/4).evalf(subs={'t': t1/sqrt(2)}), adaptive=False, nb_of_points=1000)
plt.show()
```

```python
plot((f2/(2*sqrt(2))).evalf(subs={'t': t1/sqrt(2)}), adaptive=False, nb_of_points=1000)
plt.show()
```

```python
p = plot(f.evalf(subs={'t':t1}), adaptive=False, nb_of_points=1000, show=False, label="F(t1)")
p.extend(plot((f3/4).evalf(subs={'t': t1/sqrt(2)}), adaptive=False, nb_of_points=1000, show=False, label="beta(t1)"))
p.extend(plot((f2/(2*sqrt(2))).evalf(subs={'t': t1/sqrt(2)}), adaptive=False, nb_of_points=1000, show=False, label="alpha(t1)"))
p.legend = True
p.show()
```
