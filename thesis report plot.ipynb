```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import math
```

```python
def f(x):
    return np.sin(4*np.pi*x)
```

```python
np.random.seed(42)
x = np.linspace(0, 1, 1000).round(5)
y = f(x) + np.random.normal(scale=0.3, size=(1000,))
```

```python
plt.figure(figsize=(21, 9), dpi=300)
plt.scatter(x, y, s=10)
plt.plot(x, f(x), c="k", linewidth=2)
plt.vlines(x=np.arange(start=1, stop=50)/50, ymin=-1.6, ymax=1.6, linestyles="dashed", colors="lime")
plt.grid()
plt.show()
```
